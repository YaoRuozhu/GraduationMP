```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sample data representing the mean and standard deviation for each activity
activities = ['A', 'B', 'C', 'D']
activity_means = [0.2, 0.3, 0.4, 0.5]  # Mean values for each activity
activity_stdevs = [0.1, 0.2, 0.3, 0.1]  # Standard deviations for each activity

# Generate x-axis values (activities)
x = np.linspace(0, len(activities), 100)

# Plot the normal distribution for each activity
for i, (mean, stdev) in enumerate(zip(activity_means, activity_stdevs)):
    y = np.exp(-(x - i) ** 2 / (2 * stdev ** 2)) / (stdev * np.sqrt(2 * np.pi))
    plt.plot(x, y, label=activities[i])

plt.xlabel('Activity')
plt.ylabel('Probability')
plt.title('Normal Distribution: Next Activity Prediction')
plt.legend()
plt.show()
```

```python
# Define the activities and their probabilities
activities = ['A', 'B', 'C', 'D']
probabilities = [0.2, 0.3, 0.25, 0.25]  # Probabilities of each activity

# Calculate the mean and standard deviation of the normal distribution
mean = np.dot(np.arange(len(activities)), probabilities)
std = np.sqrt(np.dot((np.arange(len(activities)) - mean) ** 2, probabilities))

# Generate the x-axis values
x = np.arange(len(activities))

# Calculate the PDF values for each x-axis value
pdf = 1 / (std * np.sqrt(2 * np.pi)) * np.exp(-(x - mean)**2 / (2 * std**2))

# Plot the normal distribution figure
plt.bar(x, pdf)
plt.xlabel('Activity')
plt.ylabel('Probability')
plt.title('Normal Distribution: Next Activity Prediction')
plt.xticks(x, activities)
plt.show()
```
